## Part 1 - Personal Finance Planner

### Collect Crypto Prices Using the `requests` Library

```python
# Set current amount of crypto assets
my_btc = 1.2
my_eth = 5.3
```

```python
# Crypto API URLs
btc_url = "https://api.alternative.me/v2/ticker/Bitcoin/?convert=CAD"
eth_url = "https://api.alternative.me/v2/ticker/Ethereum/?convert=CAD"
```

```python
# Fetch current BTC price
btc_info = requests.get(btc_url)
btc_data = btc_info.json()
btc_price = btc_data["data"]["1"]["quotes"]["CAD"]["price"]

# Fetch current ETH price
eth_info = requests.get(eth_url)
eth_data = eth_info.json()
eth_price = eth_data["data"]["1027"]["quotes"]["CAD"]["price"]

# Compute current value of my crpto
my_btc_value = my_btc * btc_price
my_eth_value = my_eth * eth_price

# Print current crypto wallet balance
print("The current value of your {} BTC is ${:0,.2f}".format(my_btc, my_btc_value))
print("The current value of your {} ETH is ${:0,.2f}".format(my_eth, my_eth_value))
```

### Collect Investments Data Using Alpaca: `SPY` (stocks) and `AGG` (bonds)

```python
# Set current amount of shares
my_agg = 200
my_spy = 50
```

```python
# Set Alpaca API key and secret
alpaca_api_key = os.getenv("ALPACA_API_KEY")
alpaca_secret_key = os.getenv("ALPACA_SECRET_KEY")

# Create the Alpaca API object
alpaca = tradeapi.REST(
    alpaca_api_key,
    alpaca_secret_key,
    api_version="v2")
```

```python
# Format current date as ISO format
today = pd.Timestamp("2022-09-16", tz="America/New_York").isoformat()

# Set the tickers
tickers = ["AGG", "SPY"]

# Set timeframe to "1Day" for Alpaca API
timeframe = "1Day"

# Get current closing prices for SPY and AGG
df_portfolio = alpaca.get_bars(
    tickers,
    timeframe,
    start = today,
    end = today
).df

# Reorganize the DataFrame
# Separate ticker data
AGG = df_portfolio[df_portfolio['symbol']=='AGG'].drop('symbol', axis=1)
SPY = df_portfolio[df_portfolio['symbol']=='SPY'].drop('symbol', axis=1)


# Concatenate the ticker DataFrames
df_portfolio = pd.concat([AGG, SPY],axis=1, keys=['AGG','SPY'])

# Preview DataFrame
df_portfolio
```

```python
# Pick AGG and SPY close prices
agg_close_price = float(df_portfolio["AGG"]["close"])
spy_close_price = float(df_portfolio["SPY"]["close"])


# Print AGG and SPY close prices
print(f"Current AGG closing price: ${agg_close_price}")
print(f"Current SPY closing price: ${spy_close_price}")
```

```python
# Compute the current value of shares
my_agg_value = my_agg * agg_close_price
my_spy_value = my_spy * spy_close_price

# Print current value of shares
print("The current value of your {} AGG shares is ${:0,.2f}".format(my_agg, my_agg_value))
print("The current value of your {} SPY shares is ${:0,.2f}".format(my_spy, my_spy_value))
```

### Savings Health Analysis

```python
# Set monthly household income
monthly_income = 12000

# Consolidate financial assets data
crypto_value = round(my_btc_value + my_eth_value,2)
shares_value = round(my_agg_value + my_spy_value,2)

# Create savings DataFrame

assets = ['crypto', 'shares']
values = [crypto_value, shares_value]

df_savings=pd.DataFrame(data={'amount': values}, index=assets)

# Display savings DataFrame
display(df_savings)
```

```python
# Plot savings pie chart
df_savings.plot.pie(y="amount", title="Personal Savings Composition", figsize=(10,20))
```

```python
# Set ideal emergency fund
emergency_fund = monthly_income * 3

# Calculate total amount of savings
total_savings = crypto_value + shares_value

# Validate saving health
if (total_savings > emergency_fund):
    print("Congratulations!! Your total savings is ${:0,.2f} and you have enough money for a ${:0,.2f} emergency fund.".format(total_savings, emergency_fund))
elif (total_savings == emergency_fund):
    print("Congratulations on reaching your financial goal!! Your total savings is ${:0,.2f} and you have sufficient money for a ${:0,.2f} emergency fund.".format(total_savings, emergency_fund))
else:
    print("Sorry you have NOT reached your financial goal. Your have a total savings of ${:0,.2f} and are ${:0,.2f} away from reaching your ${:0,.2f} emergency fund goal.".format(total_savings, (emergency_fund-total_savings), emergency_fund)) 
```

## Part 2 - Retirement Planning

### Monte Carlo Simulation

```python
# Set start and end dates of five years back from today.
# Sample results may vary from the solution based on the time frame chosen
start_date = pd.Timestamp('2016-05-01', tz='America/New_York').isoformat()
end_date = pd.Timestamp('2021-05-01', tz='America/New_York').isoformat()
```

```python
# Get 5 years' worth of historical data for SPY and AGG
df_portfolio_data = alpaca.get_bars(
    tickers,
    timeframe,
    start=start_date,
    end=end_date
).df

# Reorganize the DataFrame
# Separate ticker data
AGG_bond = df_portfolio_data[df_portfolio_data['symbol']=='AGG'].drop('symbol', axis=1)
SPY_stock = df_portfolio_data[df_portfolio_data['symbol']=='SPY'].drop('symbol', axis=1)


# Concatenate the ticker DataFrames
df_portfolio_data = pd.concat([AGG_bond, SPY_stock], axis=1, keys=["AGG", "SPY"])

# Display sample data
df_portfolio_data.head()
```

```python
# Configuring a Monte Carlo simulation to forecast 30 years cumulative returns
MC_30year = MCSimulation(
    portfolio_data = df_portfolio_data,
    weights = [0.40,.60],
    num_simulation = 500,
    num_trading_days = 252*30
)
```

```python
# Printing the simulation input data
MC_30year.portfolio_data.head()
```

```python
# Running a Monte Carlo simulation to forecast 30 years cumulative returns
MC_30year.calc_cumulative_return()
```

```python
# Plot simulation outcomes
line_plot = MC_30year.plot_simulation()

# Save the plot for future usage
line_plot.get_figure().savefig("MC_30year_sim_plot.png", bbox_inches="tight")
```

```python
# Plot probability distribution and confidence intervals
dist_plot = MC_30year.plot_distribution()
```

### Retirement Analysis

```python
# Fetch summary statistics from the Monte Carlo simulation results
statistics = MC_30year.summarize_cumulative_return()

# Print summary statistics
print(statistics)
```

### Calculate the expected portfolio return at the `95%` lower and upper confidence intervals based on a `$20,000` initial investment.

```python
# Set initial investment
initial_investment = 20000

# Use the lower and upper `95%` confidence intervals to calculate the range of the possible outcomes of our $20,000
ci_lower = round(statistics[8]*initial_investment,2)
ci_upper = round(statistics[9]*initial_investment,2)

# Print results
print("There is a 95% chance that an initial investment of ${:0,.2f} in the portfolio"
      " over the next 30 years will end within in the range of"
      " ${:0,.2f} and ${:0,.2f}".format(initial_investment, ci_lower, ci_upper))
```

### Calculate the expected portfolio return at the `95%` lower and upper confidence intervals based on a `50%` increase in the initial investment.

```python
# Set initial investment
initial_investment = 20000 * 1.5

# Use the lower and upper `95%` confidence intervals to calculate the range of the possible outcomes of our $30,000
ci_lower = round(statistics[8]*initial_investment,2)
ci_upper = round(statistics[9]*initial_investment,2)

# Print results
print("There is a 95% chance that an initial investment of ${:0,.2f} in the portfolio"
      " over the next 30 years will end within in the range of"
      " ${:0,.2f} and ${:0,.2f}".format(initial_investment, ci_lower, ci_upper))
```

### Although the investor has a potentially greater return (upper CI portfolio value of 227k) by holding more stocks in their portfolio, they also have a greater risk in that their is also a 5% chance that the investor could end up losing 25k of their initial investment (lower CI portfolio value of 35k).

### It should also be noted that the Monte Carlo Simulation analysis above is by no means a guarentee of returns. The purpose is to get a better understanding the risk of the investment. In this case by holding more stock, their investor increases their chances of a higher return BUT also increases their risk profile.

### With the above in mind, if would premature to conclude that the credit union members would be able to retire after 10years by simply holding more stocks in their portfolio. Instead, the portfolio weighting of stocks vs. bonds would need to monitored and managed on an ongoing babsis in order for the investor to to meet their financial needs.
